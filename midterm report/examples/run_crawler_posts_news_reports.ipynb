```python
# 测试获取股吧标题
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import re
import time

# 配置浏览器（使用webdriver-manager自动管理驱动）
service = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)

# 要爬取的股票列表（示例：包含600519及其他股票）
#select_stocks = ('600519', '600036', '000858', '300750')
select_stocks = ('600519', '600036')

# 循环爬取每个股票的股吧标题
for stock in select_stocks:
    print(f"\n===== 开始爬取股票 {stock} 的股吧标题 =====")
    # 构造股吧URL（东方财富网股吧格式）
    url = f'http://guba.eastmoney.com/list,{stock}_2.html'
    
    try:
        # 访问页面
        browser.get(url)
        time.sleep(2)  # 等待页面加载
        
        # 获取页面源码
        data = browser.page_source
        
        # 正则提取标题（匹配title属性中的内容）
        p_title = '<a href=".*?" title="(.*?)"'  # 提取title属性值
        titles = re.findall(p_title, data, re.S)  # re.S确保.匹配换行符
        
        # 打印结果
        if titles:
            for i, title in enumerate(titles, 1):
                print(f"{i}. {title}")
            print(f"共提取到 {len(titles)} 条标题")
        else:
            print("未提取到任何标题")
    
    except Exception as e:
        print(f"爬取股票 {stock} 失败：{str(e)}")
    
    # 间隔1秒，避免请求过快
    time.sleep(1)

# 关闭浏览器
browser.quit()
```

```python
# 测试获取咨讯信息（标题、网址、时间）
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import re
import time

# 配置浏览器（使用webdriver-manager自动管理驱动）
service = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)

# 要爬取的股票列表（示例：包含600519及其他股票）
#select_stocks = ('600519', '600036', '000858', '300750')
select_stocks = ("招商银行", "紫金矿业")

# 循环爬取每个股票的股吧标题
for stock_name in select_stocks:
    print(f"\n===== 开始爬取股票 {stock_name} 的新闻信息 =====")
    # 构造股吧URL（东方财富网股吧格式）
    url = f'http://so.eastmoney.com/news/s?keyword={stock_name}'
    
    try:
        browser.get(url)
        time.sleep(5)  # 等待页面加载
        data = browser.page_source

        p_title = '<div class="news_item_t".*?><a href=".*?" target="_blank">(.*?)</a></div>'
        p_date = '<div class="news_item_c"><span class="news_item_time">(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) - </span>'
        p_href = '<div class="news_item_url".*?><a href="(.*?)" target="_blank">.*?</a></div>'
        title = re.findall(p_title, data, re.S | re.M)
        title = [re.sub(r'\s+', ' ', t).strip() for t in title]
        title = [re.sub(r'</?em>', '', s) for s in title]
        href = re.findall(p_href, data, re.S)
        date = re.findall(p_date, data)
        
        # 打印结果
        if title:
            for i in range(len(title)):
                title[i] = re.sub('<.*?>', '', title[i])
                print(str(i+1) + '.' + title[i] + ' - '+ date[i])
                print(href[i])
        else:
            print("未提取到任何标题")
    
    except Exception as e:
        print(f"爬取股票 {stock_name} 失败：{str(e)}")
    
    # 间隔1秒，避免请求过快
    time.sleep(1)

# 关闭浏览器
browser.quit()
```

```python
# 测试获取研报pdf
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import re
import time
import requests
import os

# 配置浏览器（使用webdriver-manager自动管理驱动）
service = Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=service)

# 要爬取的股票列表
#select_stocks = ('600519', '600036', '000858', '300750')
select_stocks = ("招商银行", "五粮液")

# 循环爬取每个股票的研报pdf
for stock_name in select_stocks:
    print(f"\n===== 开始爬取股票 {stock_name} 的研报pdf =====")

    url = f'http://so.eastmoney.com/Yanbao/s?keyword={stock_name}'
    browser.get(url)
    time.sleep(2)  # 等待页面加载完成
    data = browser.page_source

    p_href = '<div class="notice_item_t".*?>.*?<a href="(.*?)"'
    href = re.findall(p_href, data)

    for i in range(len(href)):
        # 1.请求研报跳转页面
        browser.get(href[i])
        time.sleep(2)  # 等待页面加载完成
        data = browser.page_source
        
        # 2.提取跳转页面中的标题和PDF下载网址
        p_name = '<h1 id="zw-title">(.*?)</h1>'
        p_href_pdf = '<a class="pdf-link" href="(.*?)">'
        name = re.findall(p_name, data)
        href_pdf = re.findall(p_href_pdf, data)[0]
        href_pdf = href_pdf.split('?')[0]
        
        # 3.利用requests下载PDF文件
        dir_path = stock_name # 目录名称
        file_name = f"{name[0]}.pdf"  # 文件名
        path = os.path.join(dir_path, file_name)  # 组合目录和文件名
        if not os.path.exists(dir_path):
            os.makedirs(dir_path)  # 递归创建目录（如果父目录不存在也会创建）
                                                                                                                                                                                                                                                                                                               
        try:
            res = requests.get(href_pdf, timeout=30)  # 增加超时时间，避免无限等待
            res.raise_for_status()  # 检查请求是否成功（状态码200）
            with open(path, 'wb') as file:  # 使用with语句自动关闭文件，更安全
                file.write(res.content)
            print(f"PDF保存成功：{path}")
        except Exception as e:
            print(f"下载或保存失败：{str(e)}")

# 关闭浏览器
browser.quit()
```
