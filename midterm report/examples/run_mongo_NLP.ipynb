```python
import paddlehub as hub
from mongodb import MongoAPI
import pandas as pd

# 从MongoDB中读取指定股票的post_info数据
stock_code = '000333'  # 可修改为其他股票代码
postdb = MongoAPI('post_info', f'post_{stock_code}')
post_docs = list(postdb.find({}, {'post_title': 1, 'post_date': 1, 'post_time': 1}))
data_raw = pd.DataFrame(post_docs)

# 重命名列以匹配后续处理
data_raw.rename(columns={'post_title': 'title'}, inplace=True)

# 将post_date和post_time连接成完整时间戳
if 'post_date' in data_raw.columns and 'post_time' in data_raw.columns:
    data_raw['time'] = pd.to_datetime(data_raw['post_date'] + ' ' + data_raw['post_time'].astype(str))
else:
    data_raw['time'] = pd.to_datetime('2020-01-01')

##使用本地下载的百度开源的SKEP模型来预测情感倾向
senta = hub.Module(directory="ernie_skep_sentiment_analysis")
texts = data_raw['title'].tolist()
res = senta.predict_sentiment(texts,use_gpu=False)
data_raw['pos_p'] = [x['positive_probs'] for x in res]
##重采样至十五分钟
data_raw.index = data_raw['time']
data = data_raw.resample('15min').mean().reset_index()
```

```python
# 导入库与初始化配置
import jieba
import jieba.posseg as pseg
import re
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 自定义金融领域停用词表
finance_stopwords = set([
    '的', '了', '在', '是', '和', '到', '一', '个', '为', '与', '从', '这', '那', '会',
    '把', '被', '比', '将', '于', '上', '下', '中', '有', '没', '但', '要', '也', '还',
    '很', '更', '最', '或', '及', '其他', '等', '所', '能', '可', '不', '只', '如果',
    '因为', '所以', '如今', '现在', '今天', '明天', '昨天', '周一', '周二', '周三', '周四', '周五',
    '月', '日', '年', '号', '点', '分', '秒', '据', '来源', '原文', '记者', '编辑', '发布',
    '网友', '用户', '投资者', '分析师', '专家'
])

print("初始化完成！")
```

```python
# 文本预处理函数定义
def preprocess_text(text):
    """清洗文本：移除特殊字符和数字"""
    text = re.sub(r'[^\u4e00-\u9fff\w]', '', str(text))
    return text

def tokenize_and_filter(text):
    """分词、去停用词、词性过滤"""
    # 清洗文本
    text = preprocess_text(text)
    
    # 分词并进行词性标注
    words_pos = pseg.cut(text)
    
    # 过滤停用词和指定词性
    filtered_words = []
    for word, flag in words_pos:
        if word not in finance_stopwords and len(word) > 1 and flag not in ['x', 'u', 'c']:
            filtered_words.append(word)
    
    return filtered_words

import paddlehub as hub
from mongodb import MongoAPI
import pandas as pd

# 从MongoDB中读取指定股票的post_info数据
stock_code = '000333'  # 可修改为其他股票代码
postdb = MongoAPI('post_info', f'post_{stock_code}')
post_docs = list(postdb.find({}, {'post_title': 1, 'post_date': 1, 'post_time': 1}))
data_raw = pd.DataFrame(post_docs)

# 重命名列以匹配后续处理
data_raw.rename(columns={'post_title': 'title'}, inplace=True)

# 将post_date和post_time连接成完整时间戳
if 'post_date' in data_raw.columns and 'post_time' in data_raw.columns:
    data_raw['time'] = pd.to_datetime(data_raw['post_date'] + ' ' + data_raw['post_time'].astype(str))
else:
    data_raw['time'] = pd.to_datetime('2020-01-01')

# 对data_raw进行文本预处理
print("开始文本预处理...")
data_raw['tokens'] = data_raw['title'].apply(tokenize_and_filter)
data_raw['tokens_text'] = data_raw['tokens'].apply(lambda x: ' '.join(x))

print("✓ 文本预处理完成！")
print(f"  处理文本数：{len(data_raw)} 条")
print(f"  样本分词结果：{data_raw['tokens'].iloc[0][:5]}...")
```

```python
# 高频关键词统计与词云可视化
print("生成词云...")

# 合并所有分词结果
all_tokens = []
for tokens in data_raw['tokens']:
    all_tokens.extend(tokens)

# 统计词频
word_freq = Counter(all_tokens)
top_words = word_freq.most_common(50)

print(f"✓ 高频关键词统计完成（共 {len(word_freq)} 个不同词汇）")
print(f"\n  Top 10 高频词汇：")
for i, (word, count) in enumerate(top_words[:10], 1):
    print(f"    {i:2d}. {word}: {count:4d} 次")

# 生成词云图
wordcloud_text = ' '.join(all_tokens)
wordcloud = WordCloud(
    font_path='C:\\Windows\\Fonts\\msyh.ttc',
    width=800,
    height=600,
    background_color='white',
    colormap='viridis',
    max_words=100
).generate(wordcloud_text)

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'股票 {stock_code} 舆情高频关键词词云', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("✓ 词云可视化完成！")
```

```python
# TF-IDF特征提取与K-Means主题聚类
print("进行主题聚类...")

# 过滤掉过短的文本
data_clean = data_raw[data_raw['tokens_text'].str.len() > 0].copy()

# TF-IDF特征提取
tfidf_vectorizer = TfidfVectorizer(
    max_features=200,
    min_df=2,
    max_df=0.8,
    ngram_range=(1, 2)
)
tfidf_features = tfidf_vectorizer.fit_transform(data_clean['tokens_text'])

# K-Means聚类（簇数k=5）
k = min(5, len(data_clean))  # 确保k不超过数据量
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
data_clean['topic_cluster'] = kmeans.fit_predict(tfidf_features)

print(f"✓ 聚类完成！")
print(f"  簇数：{k}")
print(f"  文本特征维度：{tfidf_features.shape}")
print(f"\n  各主题数量分布：")
topic_dist = data_clean['topic_cluster'].value_counts().sort_index()
for topic_id, count in topic_dist.items():
    print(f"    主题 {topic_id}：{count:4d} 条（{count/len(data_clean)*100:5.1f}%）")
```

```python
# 聚类结果可视化
print("绘制聚类分析图表...")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 主题分布饼图
topic_counts = data_clean['topic_cluster'].value_counts().sort_index()
colors = plt.cm.Set3(np.linspace(0, 1, len(topic_counts)))
axes[0].pie(topic_counts.values, labels=[f'主题{i}' for i in topic_counts.index], 
            autopct='%1.1f%%', colors=colors, startangle=90)
axes[0].set_title('舆情主题聚类分布', fontsize=14, fontweight='bold')

# 各主题的平均情感倾向柱状图
data_clean['pos_p'] = data_raw.loc[data_clean.index, 'pos_p']
topic_sentiment = data_clean.groupby('topic_cluster')['pos_p'].mean()
bars = axes[1].bar(range(len(topic_sentiment)), topic_sentiment.values, color=colors, edgecolor='black', linewidth=1.5)
axes[1].set_xlabel('主题编号', fontsize=12)
axes[1].set_ylabel('平均正面概率', fontsize=12)
axes[1].set_title('各主题平均情感倾向', fontsize=14, fontweight='bold')
axes[1].set_xticks(range(len(topic_sentiment)))
axes[1].set_xticklabels([f'主题{i}' for i in topic_sentiment.index])
axes[1].set_ylim([0, 1])
axes[1].grid(axis='y', alpha=0.3, linestyle='--')

# 在柱子上标注数值
for bar in bars:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.3f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

print("✓ 可视化完成！")
```

```python
# 输出各主题的关键词特征
print("各主题的核心关键词：\n")

feature_names = np.array(tfidf_vectorizer.get_feature_names_out())
for topic_id in range(k):
    # 获取该主题的中心向量中最重要的特征
    center = kmeans.cluster_centers_[topic_id]
    top_indices = center.argsort()[-10:][::-1]
    top_features = feature_names[top_indices]
    
    print(f"主题 {topic_id}:")
    print(f"  关键词：{', '.join(top_features)}")
    print()

print("✓ 分析完成！")
```
